## 1. Line Plot

```python
import matplotlib.pyplot as plt

#Sample data
months=['Jan','Feb','Mar', 'April', 'May', 'June']
energy_consumption = [1200,1300,1100,1500,1400,1600]

# plot a line plt
plt.plot(months,energy_consumption, marker='o', linestyle='--')
plt.xlabel('Months')
plt.ylabel('Energy Consumption (mWh)')
plt.title('Energy Consumption in 6 month')
plt.show()
```

## 2.  Bar Chart

```python
energy_sources = ['Solar','Wind','Hydropower','Biomass']
energy_values = [1200,3400,2900,2500]
plt.bar(energy_sources,energy_values, color='lightgreen')
plt.title('Energy consumption by Renewable Energy Sources')
plt.xlabel('Energy Sources')
plt.ylabel('Energy values (MWh)')
plt.show()
```

```python
## 3. Pie Chart
```

```python
# to show energy consumption distribution
plt.pie(energy_values, labels=energy_sources, autopct='%1.1f%%', colors=['gold','gray','coral','pink'])
plt.title('Energy Consumption shares by sources')
plt.show()
```

##  4. Scatter Plot

```python
carbon_emissions =[200,500,450,300]  #kg CO2
# Create Scatter Plot
plt.scatter(energy_values, carbon_emissions, color='red')
plt.title("Energy Consumption vs Carbon Emission")
plt.xlabel('Energy Consumption (MWh)')
plt.ylabel('Carbon Emission (kgCO2)')
plt.show()
```

## 5.  Visualization - Customization

```python
#create a bar chart
plt.bar(energy_sources,energy_values, color='orange', edgecolor='black', linewidth=2)
plt.title("Customized Energy Consumption by Source", fontsize=14, fontweight='bold')
plt.xlabel('Energy Source', fontsize=11)
plt.ylabel('Energy Consumption(MWh)', fontsize=11)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(['Energy Consumption'], loc='upper right')
plt.show()
```
