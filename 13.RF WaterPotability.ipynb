```python
# Libraries
import pandas as pd
import numpy as npp
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# load the dataset 
df = pd.read_csv('water_potability.csv')
df.head()
```

```python
df.shape
```

```python
df['Potability'].unique()
```

```python
df['Potability'].nunique()
```

```python
# To check missing values
df.isnull().sum()
```

```python
# Data Exploration
df.info()
```

```python
# Statistics of data
df.describe()
```

```python
# handle the missing values - impute - mean
df1 =df.fillna(df.mean())    # or df.fillna(df.mean(), unplace=True)
df1.isnull().sum()
```

```python
# Feature selestion
X = df.drop('Potability',axis=1)
y = df['Potability']
X.shape
```

```python
X.columns
```

```python
# Split into train and test sets
X_tr, X_te, y_tr, y_te = train_test_split(X,y, train_size=0.8, random_state=42)
```

```python
X_tr.shape
```

```python
# Initialize and train
rfc =RandomForestClassifier(random_state=42)
rfc.fit(X_tr,y_tr)
```

```python
# Prediction
y_pred = rfc.predict(X_te)
y_pred
```

```python
# Evaluate the model
acc =accuracy_score(y_te, y_pred)
print(f'Accuracy Score is {acc: .2f}')
print('Classificaation Report')
print(classification_report(y_te, y_pred))
```

```python
# Confusion matrix
cm = confusion_matrix(y_te, y_pred)
sns.heatmap(cm, annot=True,fmt='d', cmap='Oranges',
           xticklabels=['Unsafe','Safe'],
           yticklabels=['Unsafe','Safe'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
# Feature Importance
importances = rfc.feature_importances_
features = X.columns
indices = npp.argsort(importances)
```

```python
# plot 
plt.figure(figsize=(10,6))
plt.title('Feature Impotances')
plt.barh(range(len(indices)), importances[indices], align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```
